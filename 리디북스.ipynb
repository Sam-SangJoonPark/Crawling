```python
import pandas as pd
import requests
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import random
import time
from tqdm import tqdm, trange
import re
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
url ='https://ridibooks.com/'
driver.get(url)
```

```python
df = pd.read_csv(r'C:\Users\mycel\Downloads\res.csv')
df
```

```python
ridi_url = {'title' : 'url'}
for i in range(len(df)) :
    try :
        driver.get(url)
        time.sleep(2)
        driver.find_element_by_xpath('//*[@id="__next"]/div/header/nav/div/form/div/div[1]/label/input').send_keys(df['title'].iloc[i])
        time.sleep(1)
        elem = driver.find_element_by_xpath('//*[@id="__next"]/div/header/nav/div/form/div/div[1]/label/input')
        time.sleep(1)
        elem.send_keys(Keys.ENTER)
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="__next"]/main/section/ul[2]/li[1]/div/a').click()
        time.sleep(2)
        book_info = {df['title'].iloc[i] : driver.current_url}
        ridi_url.update(book_info)
    except :
        try :
            driver.get(url)
            time.sleep(2)
            driver.find_element_by_xpath('//*[@id="__next"]/div/header/nav/div/form/div/div[1]/label/input').send_keys(df['title'].iloc[i] + ' ' + df['publisher'].iloc[i])
            time.sleep(1)
            elem = driver.find_element_by_xpath('//*[@id="__next"]/div/header/nav/div/form/div/div[1]/label/input')
            time.sleep(1)
            elem.send_keys(Keys.ENTER)
            time.sleep(1)
            driver.find_element_by_xpath('//*[@id="__next"]/main/section/ul/li/div/a/div').click()
            time.sleep(2)
            book_info = {df['title'].iloc[i] : driver.current_url}
            ridi_url.update(book_info)
        except :
            print('제목 : ' , df['title'].iloc[i],' ', i, ' 번 째 ')
            pass
```

```python
for i in ridi_url.keys() :
    for a in range(len(df)) :
        if i == df['title'].iloc[a] :
            df['url2'].iloc[a] = ridi_url[i]
```
