```python
# page_source 받아오기!
source = chrome.page_source
print(source)
from bs4 import BeautifulSoup
soup = BeautifulSoup(source, 'html.parser')
```

```python
# 저번에 만든 메뉴 가져오기 사용
meals = soup.find_all(class_='tch-lnc-wrap')
dic = {}
for meal in meals :
    dic.update({meal.dt.text : meal.ul.text})
print(dic)
```
