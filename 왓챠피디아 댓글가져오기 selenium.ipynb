```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
main_url = 'https://pedia.watcha.com/ko-KR'
driver = webdriver.Chrome()
driver.get(main_url)
time.sleep(3)
# 검색창 찾기
elem = driver.find_element_by_css_selector('#global_search_input')
# 검색창에 입력
elem.send_keys('박쥐')
elem.send_keys(Keys.ENTER)
time.sleep(3)
# 첫 번째 영화 클릭
try :
    elem = driver.find_element_by_css_selector('#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > section > section:nth-child(2) > div.css-wyngcf-ScrollBarContainer.e1f5xhlb0 > div.css-chidac-ScrollBar.e1f5xhlb1 > div > div > div > ul > li:nth-child(1) > a > div.css-18zdrlt-Info.e32z22d5 > div.css-1t67tr2-Titles.e32z22d4 > div.css-1q4vy80-ResultTitle.eldorhe1')
    elem.click()
    time.sleep(1)
# try 부분으로 안 되는 영화도 있음.
except :
    elem = driver.find_element_by_css_selector('#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > section > section:nth-child(2) > div.css-wyngcf-ScrollBarContainer.e1f5xhlb0 > div.css-chidac-ScrollBar.e1f5xhlb1 > div > div > div > ul > li > a > div.css-qo7iw9-ImageBlock.e32z22d3 > div > span')
    elem.click()
    time.sleep(1)
    # 커멘트 더보기 버튼 클릭  -  4번째일 때도 있고, 5번째일 때도 있음. 영화별로 상이.
    
# 일시중지
input()
    
try :
    elem = driver.find_element_by_css_selector('#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > div > div > div > div > div > div:nth-child(1) > div.css-4wqbsn-MaxWidthCol.e1pdhzq90 > div > div > section:nth-child(4) > div:nth-child(1) > div > header > div > div > a')
except :
    elem = driver.find_element_by_css_selector('#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > div > div > div > div > div > div:nth-child(1) > div.css-4wqbsn-MaxWidthCol.e1pdhzq90 > div > div > section:nth-child(5) > div:nth-child(1) > div > header > div > div > a')
elem.click()
# 무한 스크롤
import time
SCROLL_PAUSE_TIME = 3
# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait to load page
    time.sleep(1)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")
    time.sleep(SCROLL_PAUSE_TIME)
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
time.sleep(2)
# 댓글 하나씩 가져오기
i = 1
while True :
    try :
        try :
                # 스포일러 버튼 보기
                elem = driver.find_element_by_css_selector('#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > section > div > div > div > ul > div:nth-child(%d) > div.css-103yoki-TextBlock.e1oskw6f3 > span > button' % i)
                elem.click()
                time.sleep(0.2)
                elem = driver.find_element_by_css_selector('#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > section > div > div > div > ul > div:nth-child(%d) > div.css-103yoki-TextBlock.e1oskw6f3 > div > span' % i)
                print('*********',' 댓글 ',i,'****** 스포있음')
                print(elem.text)
                print('******************************')
                i+=1

        except :
            # 스포일러 없는 댓글 바로 가져오기
            elem = driver.find_element_by_css_selector('#root > div > div.css-1sh3zvx-NavContainer.ebsyszu0 > section > section > div > div > div > ul > div:nth-child(%d) > div.css-103yoki-TextBlock.e1oskw6f3 > div' % i)
            print('*********',' 댓글 ',i,'****** 스포없음')
            print(elem.text)
            print('******************************')
            i+=1           
    except :
        print('------------댓글 끝--------------')
        break
```
